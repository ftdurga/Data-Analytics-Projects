Every year, American high school students take SATs, which are standardized tests intended to measure literacy, numeracy, and writing skills. There are three sections - reading, math, and writing, each with a **maximum score of 800 points**. These tests are extremely important for students and colleges, as they play a pivotal role in the admissions process.

Analyzing the performance of schools is important for a variety of stakeholders, including policy and education professionals, researchers, government, and even parents considering which school their children should attend. 

You have been provided with a dataset called `schools.csv`, which is previewed below.

You have been tasked with answering three key questions about New York City (NYC) public school SAT performance.

```python
import pandas as pd
```

```python
schools = pd.read_csv("schools.csv")
schools.head()
```

**Question 1: Which NYC schools have the best math results?** 

```python
# Step 1: Find the schools with an average_math score that is at least 80% of 800 (max score)
schools[(schools['average_math'] >= (0.8 * 800))]
schools.sample(10)
```

```python
# Step 2: Only display the school_name and the average_math score
schools[(schools['average_math'] >= (0.8 * 800))][['school_name', 'average_math']]
```

```python
# Step 3: Create a new dataframe for this data, and sort it by "average_math" in descending order.

best_math_schools = schools[(schools['average_math'] >= 640)][['school_name', 'average_math']]
best_math_schools = best_math_schools.sort_values('average_math', ascending=False) # RMB to reassign, else the change does not persist
best_math_schools.head()

# Alternative method that combines step 1, 2 and 3
# best_math_schools = schools[schools["average_math"] >= 640][["school_name", "average_math"]].sort_values("average_math", ascending=False)
```

**Question 2: What are the top 10 performing schools based on the combined SAT scores?**

```python
#Step 1: Create a duplicate of the original dataframe, and create a new column called 'total_sat'
schools_duplicate = schools
schools_duplicate['total_SAT'] = schools_duplicate['average_math'] + schools_duplicate['average_reading'] + schools_duplicate['average_writing']
schools_duplicate.head()
```

```python
top_10_schools = schools_duplicate[['school_name', 'total_SAT']].sort_values('total_SAT', ascending=False).head(10)
top_10_schools.head()
```

**Question 3: Which single borough has the largest standard deviation in the combined SAT score?**

```python
# Note: The function agg calculates the statistics based on the coloumn 'total_SAT'
# .agg(["count", "mean", "std"]) → calculates:
    # count → number of schools in each borough
    # mean → average SAT score
    # std → standard deviation (how spread out scores are)

boroughs = schools.groupby("borough")["total_SAT"].agg(["count", "mean", "std"]).round(2)
```

```python
# Filter for max std based on std, note: there can be more than 1 borough with the max std
largest_std_dev = boroughs[boroughs["std"] == boroughs["std"].max()]
largest_std_dev.head()
```

```python
# Rename the columns for clarity
largest_std_dev = largest_std_dev.rename(columns={"count": "num_schools", "mean": "average_SAT", "std": "std_SAT"})
largest_std_dev.head()
```

```python
# Move borough from index to column
largest_std_dev.reset_index(inplace=True)
largest_std_dev.head()
```
